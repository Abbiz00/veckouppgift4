
1g Funktionen "isinstance" kan kontrollera en variabels datatyp. Vad gör funktionen is_number? Går det att förbättra koden? Funktionen svara True på både int och float


```
def is_number(x):
    if isinstance(x, int):
        return True
    elif isinstance(x, float):
        return True
    return False

print(is_number(5.5))
print(is_number(42))
```

Delat upp i två funktioner en för kontroll av int eller float.

```
def is_number_int(x):
    return isinstance(x, int)   #isinstance ger en True eller false så man behöver inte if satsen

def is_number_float(x):
    return isinstance(x, float)

print(f"Är 5.5 en int? {is_number_int(5.5)}")
print(f"Är 42 en int? {is_number_int(42)}")
print("")
print(f"Är 5.5 en float? {is_number_float(5.5)}")
print(f"Är 42 en float? {is_number_float(42)}")
```

```
def average_words(strings):
    found = []
    for item in strings:
        if 4 < len(item) < 8:
            found.append(item)
    return found

average_words(["sup", "how's", "it", "going", "reflecting", "on", "programs", "and", "coding","333", "4444", "55555", "666666", "7777777", "88888888", "99999999" ])
```

```
def length_words(strings):
    found = []
    for item in strings:
        if 5 <= len(item) <= 7:
            found.append(item)
    return found

length_words(["sup", "how's", "it", "going", "reflecting", "on", "programs", "and", "coding","333", "4444", "55555", "666666", "7777777", "88888888", "99999999" ])
```

1i En uppgift i tre delar:
Lista ut vad som är funktionens syfte, baserad på namn och sammanhang.
1) Funktionen skall räkna ut det minsta talet i listan som skickas in
2) Förväntade svaret på den första är rätt, men det är bara för att de är negativa tal


```
def find_min(numbers):
    counter = 0
    for item in numbers:
        if item < counter:
            counter = item
    print(f"The smallest item is: {counter}")
    return counter

find_min([10, 3, -4, -11])
find_min([])
find_min([100])
```

tagit bort counter = 0 som skapar problem i funktionen. Om värdena i listan är postiva så blir counter = 0

```
def find_min(numbers):
    if not numbers:
        print("The list is empty")
        return None
    smallest = numbers[0]
    for item in numbers:
        if item < smallest:
            smallest = item
    print(f"The smallest item is: {smallest}")
    return None

find_min([10, 3, -4, -11])
find_min([])
find_min([100])
```
